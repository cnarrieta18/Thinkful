### functions

```python
def unique_entries(df_column):
    count = df_column.value_counts(dropna=False)
    percentage = df_column.value_counts(normalize=True, dropna=False).mul(100).round(1).astype(str) + '%'
    df = pd.DataFrame({'count':count, 'percentage':percentage})
    df.name = df_column.name
    print(df.name)
    return df
```

```python
def unique_entries_bin(df_column,bins):
    count = df_column.value_counts(dropna=False, bins=bins)
    percentage = df_column.value_counts(normalize=True, dropna=False, bins=bins).mul(100).round(1).astype(str) + '%'
    df = pd.DataFrame({'count':count, 'percentage':percentage})
    df.name = df_column.name
    print(df.name)
    return df
```
