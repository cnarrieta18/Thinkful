There is a pattern in the null entries: Revenue and Expenditure (~14%), Grades (~9%), and Score (~64%)

- ENROLL and GRADES_ALL_G: fill with values from each other; drop remaining nulls
- REVENUE, EXPENDITURE, and GRADES: fill nulls with average for rows with AVG_SCORE entries; drop rest of nulls; mostly from non-US state label; relatively few instances
- AVG_SCORE: Fill nulls with average of the average scores

```python
useducation_test = useducation_df.copy()
```

```python
useducation_test['ENROLL'].fillna(useducation_test['GRADES_ALL_G'])
useducation_test['GRADES_ALL_G'].fillna(useducation_test['ENROLL']);
```

```python
useducation_test.dropna(subset=['TOTAL_REVENUE', 'FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE'], inplace=True)
```
